```python
DNA = 'TGGTTTTGTGCGATGTAGACTGCCGAACGCCATATCCCAGTCTCGTCATCTTTATTTGGACATGCTCCTTTCAATTAGTCGCCTAGAGCGTAATGCGCCCCTTAATCTCCTAGGCAGCCTAGATATCTATGTGGTGCACGAAGATGAATTATCGTGTTTGAGAGCTCTTCGCGCTGGTATCTTCGAACGGCAGATCTTTTGCAACCCGTGATGGCTGCAAAAAGATTTTTGGCACCAATTAGGTAAGTCCCGATATGTCCCCAGTCCAGTGGCTGTGAAGGATTAAAGCGGAGCATAGCAAGGCCCTCCCGGGACATCGTTTTCACTACCTCGGACTTGGCGTTGTAGACACGACACCGATCTTGAATCATGTAAAAACTCTGACGAGGTGCGGTTCAATGGCAGGAGTGTTTGGTCGTTTAGGACACCTCAACGCCGCAAGACTGAACCAATCCGTCATGGCTTACTATTAGCGAGGAAGCCGCACGGGAGAGCCACCCTTCATCATTCTTCGCGACGAAATCCGCTTATGTTGCTTGAGGTGACGCGCGTGACTATTAACACCTTACCCACGCCGAACTTGCCACCTGCGAGTGGTTATCACCCAAAGTACTGTCGACTGCTGGCTCATAGGTCGTGTCCTCCATCCGGCTGCTCCATATGAGCACAGGGTGACTGGGCTGGAATTAGCCGTATTGGCTAAGCTAAACTCATTCGCGAAAAAAGGTTCTAGTCTTCAATCAGGGCGTCACAACTAGGGCGTCTAGGCCACAATGACAATCAAGCTTTGATCAGTTAACAATCCAGATGAGATCCTCTTCAAACGGTTCCGAACGTGTCCCGCATAGTAGC'
```

#### Complement strand is the other strand. DNA is double stranded, and it is known that A pairs with T & G with C.

```python
DNA_Comp = DNA.replace('A' , 't')
DNA_Comp = DNA_Comp.replace('T' , 'a')
DNA_Comp = DNA_Comp.replace('C' , 'g')
DNA_Comp = DNA_Comp.replace('G' , 'c')
print(DNA_Comp)
```

#### The convention is that DNA runs from 5' to 3', and the strands are anti-parallel, hence the need for *Reverse* Complement.

```python
DNA_Comp = DNA_Comp.upper()
print(DNA_Comp)
```

```python
DNA_Comp = DNA_Comp[::-1]
print(DNA_Comp)
print(f"The complement strand sequence is {DNA_Comp}.")
```
